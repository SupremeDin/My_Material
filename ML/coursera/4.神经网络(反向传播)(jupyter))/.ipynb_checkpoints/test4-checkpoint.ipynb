```python
import numpy as np
#import pandas as pd 原数据是mat格式 不需要pandas
import matplotlib.pyplot as plt
from scipy.io import loadmat
import scipy.optimize as opt
from sklearn.metrics import classification_report #评价报告，用来计算最后的准确率
```

```python
# 1.可视化训练集

def load_mat(path):
    '''读取数据'''
    data = loadmat(path)
    X = data['X']
    y = data['y']
    return X,y

X,y = load_mat('ex4data1.mat')
X.shape,y.shape
y = y.flatten()
y.shape
```

```python
# 随机显示100张图片（100个训练样本）
def plot_100_images(X):
    index = np.random.choice(range(X.shape[0]),100)    # X.shape[0] = 5000,range(5000) = [0,1,2,...,4999]  从其中选择100
    images =X[index]  # images的维度为：（100,400）即为100张图
    #建立10*10的画布
    fig,ax = plt.subplots(10,10,sharex=True,sharey=True,figsize=(8,8))
    
    for i in range(10):
        for j in range(10):
            ax[i,j].matshow(images[10*i + j].reshape(20,20))
    plt.show()

plot_100_images(X)
```

```python
# ********** 2.数据准备 ***********

# 1.X (5000,400) ------>(5000,401) 加上x0一列
# 2.y 原数据集中y：(5000,) ------>(5000,10)【原因是one vs all，即要判断某个样本属于哪一类，要通过K个probabilitiy的max来确定
# 3.theta【weights】 在真正神经网络的时候，theta是要随机初始化值的（创建维度适合的weights，并随机初始化）。而非从数据集读取
```

```python
# 处理X
raw_X,raw_y = load_mat('ex4data1.mat')
X = np.insert(X,0,1,axis=1)  #插入之后，其维度（5000,401）
```

```python
# 处理y
def expand_y(y):    #注意到：这个y是m行1列（5000,1），而非（5000，）
    result = []
    
    for i in y:
        y_array = np.zeros(10)
        y_array[i-1] = 1
        result.append(y_array)
    
    return np.array(result)

y = expand_y(raw_y)
```

```python
# 处理theta【weights】
# 已经训练好了，现将mat类型的theta数据读取
def load_weight(path):
    data = loadmat(path)
    return data['Theta1'],data['Theta2']
theta1,theta2 = load_weight('ex4weights.mat')
# theta1.shape,theta2.shape  
# ((25, 401), (10, 26))

# 展开theta，展开成一个行向量【序列】？？？
# 将所有的权重矩阵，展开成一个行向量！（目的是什么，暂时不知道）
def serialize(a,b):    #本例中只有两个权重矩阵，故函数只有两个参数a，b
    return np.r_[a.flatten(),b.flatten()]   #具体看np.r_的用法 （此处是将两个行向量连接成一个行向量）

theta = serialize(theta1,theta2)
#theta.shape
#(10285,)   25*401 + 10*26 = 10285

# 还原theta，恢复其原来的形状
def deserialize(seq):
    return seq[:25*401].reshape(25,401),seq[25*401:].reshape(10,26)   #返回两个矩阵theta1,theta2
theta11,theta22 = deserialize(theta)
theta11.shape,type(theta22)
```

```python
# ************ 3. 函数准备 ***********
# 1.sigmoid()    【后面的cost函数用得到】
# 2.前向传播函数   【目的是得到每一层的输入z1,z2,z3和输出a1,a2,a3,其中对于第一层：z1==a1,故第一层没有误差】
# 3.cost()、regularized_cost()    【损失函数，计算两个是为了条理一点哈哈哈哈】
```

```python
# sigmoid()函数
def sigmoid(z):
    return 1/(1 + np.exp(-z))    
```

```python
# 前向传播函数
def feed_forward(theta,X):
    #传入的参数theta：是serialize后的一维向量，为什么不直接将theta1，theta2直接传入？？？
    #参数X：(5000,401)
    theta1,theta2 = deserialize(theta)   # theta1 :(25, 401)  theta2 :(10,26) 
    
    a1 = X               # a1 第一层输出值 (5000,401)
    z2 = a1@theta1.T     # z2 第二层输入值 (5000,401)*(401,25) = (5000,25)
    a2 = sigmoid(z2)     # a2 第二层输出值 (5000,25)
    a2 = np.insert(a2,0,1,axis=1)   #    (5000,26)
    z3 = a2@theta2.T     # z3 第三层输入值 (5000,26)*(26,10) = (5000,10)
    a3 = sigmoid(z3)
    
    return a1,z2,a2,z3,a3

# 现计算每一层的输入输出（由于theta、X都已知）
a1,z2,a2,z3,h = feed_forward(theta,X)     # h == a3 [:hypotheses]
#h.shape
```

```python
# cost function 
# 使用向量相乘贼简单（必须明确cost的最终结果是一个数值）
# 对应J(θ)公式：只需要知道 1.label值 y  2.预测值 h 两个矩阵的维度都是(5000,10)
def cost(theta,X,y):        #这里要传入theta，X是因为要计算 h
    a1,z2,a2,z3,h = feed_forward(theta,X)
    # 利用矩阵相乘可以快速地求
    cost = np.sum(-y*np.log(h) + (y-1)*np.log(1-h))/len(X)    # (5000,10)*(5000,10) 两个矩阵对应元素相乘，再求其sum值
    return cost
COST = cost(theta,X,y)
COST
```

```python
# Regularized cost function
#  == cost() + Reg项
# 不要将bias项进行正则化！！！！！
def regularized_cost(theta,X,y,l):       # l:正则化项系数
    theta1,theta2 = deserialize(theta)
    theta1 = theta1[:,1:]                # 将theta1，theta2 的第一列 bias项去掉
    theta2 = theta2[:,1:]
    Reg = (np.sum( theta1*theta1 ) + np.sum( theta2*theta2 ))/(2*len(X))
    return cost(theta,X,y) + Reg
costReg = regularized_cost(theta,X,y,1)
costReg
```

```python
#*************4. 反向传播 ************
# 1.sigmoid_gradient()      # 计算sigmoid的导数  【在后面的gradient的计算中会用到】
# 2.Random initialization   # 随机初始化  【打破数据的对称性；如果将theta全部初始化为0，那么将会引起？】
# 3.Back propagation (for gradient)  # 反向传播 获得gradient   【目的是获取cost function的gradient，以便在优化算法中进行求解】
# 4.regularized gradient    # 将3得到的gradient进行正则化：即：gradient + reg项 此处就是要构造这个reg项
# 5.Gradient checking       # 梯度检测    【目的是来检测通过反向传播得到的gradient是否是正确的！这个checking是通过梯度的定义求得一个矩阵（维度跟通过反向传播求得的gradient的维度是一致的）】
```

```python
# 4.1 sigmoid的梯度
def sigmoid_gradient(z):
    return sigmoid(z)*(1 - sigmoid(z))    # * 对应元素相乘
```

```python
# 4.2 随机初始化
def random_init(size):      # size == theta 他们维度一致
    '''此处的随机初始化采用：均匀分布'''
    return np.random.uniform(-0.12,0.12,size)
```

```python
# 4.3 通过反向传播 计算梯度
# 步骤：
#     1.前向传播获得各层的 输出值 a1,a2,a3 ...
#     2.反向传播获得各层的 误差值 d3,d2,d1 ...[误差为神经元的误差，故d3,d2,d1的维度跟 每层的神经元个数有关]
#     3.根据 1、2 计算的值来最终求出 每一个权重矩阵的 gradient
#     4.将所有的gradient矩阵serialize化，最终输出一维行向量（*,）的形式,且其维度与theta向量一致
def gradient(theta,X,y):     #注意：原始的梯度项 与 正则化系数项 没有任何关系
    theta1,theta2 = deserialize(theta)
    # 1
    a1,z2,a2,z3,h = feed_forward(theta,X)
    # 2
    d3 = h - y  # (5000,10) 第三层的误差
    d2 = d3@theta2[:,1:]*sigmoid_gradient(z2)  # (5000,25) 第二层的误差
    # 3
    gradient2 = d3.T@a2 # (10,26) gradient2 跟 theta2 的维度一致 
    gradient1 = d2.T@a1 # (25,401) gradient1 跟 theta1 的维度一致
    # 4
    gradient = (1/len(X))*serialize(gradient1,gradient2) # (10285,)
    
    return gradient
```

```python
# 4.4 求正则化后的 梯度
# regularized_gradient = gradient + reg项
# reg项：对theta0 不进行正则化
# theta1 : (25,401) -----> (25,400)
# theta2 : (10,26) ------> (10,25)
def regularized_gradient(theta,X,y,l):
    # 始终记住 gradient是向量！
    # 获取gradient1,gradient2
    gradient1,gradient2 = deserialize(gradient(theta,X,y))    # (25,401) (10,26)
    # 计算reg
    theta1,theta2 = deserialize(theta)
    theta1[:,0] = 0
    theta2[:,0] = 0
    reg1 = theta1
    reg2 = theta2
    # 计算 gradientReg1，gradientReg2
    gradientReg1 = gradient1 + (l/len(X))*reg1
    gradientReg2 = gradient2 + (l/len(X))*reg2

    return serialize(gradientReg1,gradientReg2)
    
    
```

```python
# 4.5 梯度检测
# 1.上面是通过反向传播 来计算 gradient
# 2.现计算`数值梯度`，又称`理论梯度` 【 根据梯度的广义定义：gradient = Δy/Δx 】
# 3.计算 gradient、numeric_gradient 之间的 relative difference【相对偏差】，若`相对偏差`很小< 10e-9,则说明之前通过bp计算的gradient是正确的
def gradient_checking(theta,X,h,e):    # 前3个参数是用bp来计算gradient，后面的e参数计算`数值梯度`
    def a_numeric_grad(plus,minus):    # plus、minus的维度跟theta的维度相同，在本例中都是(10285,)
        '''返回一个数值梯度'''
        return (regularized_cost(plus,X,y,1) - regularized_cost(minus,X,y,1))/(e*2)
    #子函数中的plus，minus跟theta的区别仅仅为：theta中的某一个特定的元素做了改动
    
    numeric_grad = [] # 先将`数值梯度`定义为一个list，后面再将其转化为array
    for i in range(len(theta)):
        # 构造plus，minus两个向量【对theta作很小改动】
        plus = theta.copy()      # deep copy otherwise you will change the raw data
        minus = theta.copy()
        plus[i] = plus[i] + e
        minus[i] = minus[i] - e
        # 计算第i个‘数值梯度’
        numeric_grad_i = a_numeric_grad(plus,minus)
        # 将所有的‘数值梯度’拼成一个 行向量
        numeric_grad.append(numeric_grad_i)
        
    numeric_grad = np.array(numeric_grad)
    analytic_grad = regularized_gradient(theta,X,y,1)
    diff = np.linalg.norm(numeric_grad - analytic_grad)/np.linalg.norm(numeric_grad + analytic_grad)
    print('If your backpropagation implementation is correct,\nthe relative difference will be smaller than 10e-9 (assume epsilon=0.0001).\nRelative Difference: {}\n'.format(diff))

gradient_checking(theta,X,y,0.0001)  
```

```python
#************* 5.学习参数 ************
# 前面完成了那么多的函数------> 只是为了计算出 regularized_gradient & regularized_cost
# 训练时候，传入参数只有 X，y (训练集，标签)
#      1. 随机初始化
#      2. 将参数 放进 minimize 函数里面
def nn_training(X,y):     
    # 随机初始化：打破数据的对称性
    init_theta = random_init(10285)    # 25*401 + 10*26
    # 调用optimize包里面的 minimize()
    res = opt.minimize(fun=regularized_cost,x0=init_theta,args=(X,y,1),method='TNC',jac=regularized_gradient,options={'maxiter':400})
    
    return res

# 训练神经网络  slow！
result = nn_training(X,y)
result
    
```

```python
# 检测学习的精度
# 使用scipy下面的classification_report

def accuracy(theta,X,y):
    _,_,_,_,h = feed_forward(theta,X)
    y_pred = np.argmax(h,axis=1) + 1
    print(classification_report(y,y_pred))
    
accuracy(result.x,X,raw_y)
```
